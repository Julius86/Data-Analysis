# NFL MVP 2024 using Data Analysis Tools

## First we import the Quarterbacks.

### Importing the first libraries we are using:

```python
from bs4 import BeautifulSoup
import requests
```

### Next we define 3 variables for the sake of making things easier. 
### The first variable "url" is the website's URL we are scrapping the data from in text form.
### The second variable "page" sends a request to the website we defined using the "url" variable we defined first.
### The "soup" variable contains a function that gets the HTML code from the website and help us utilize it for scrapping.

```python
url = 'https://www.nfl.com/stats/player-stats/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

### We define a new variable that will contain the information obtained using the method "find_all" from BeautifulSoup library. 

```python
find_headers = soup.find_all('th', class_='header')
#print(find_headers)
```

### Next we create a new variable that will contain a list with a FOR loop which will help us obtain the Headers from the table located in the website.

```python
find_header = [header.text.strip() for header in find_headers]
print(find_header)
```

### We import the Pandas library to create a new DataFrame with the information obtained from the website.

```python
import pandas as pd
```

### Creating the DataFrame using pandas.

```python
df = pd.DataFrame(columns = find_header)
df
```

### Next we create a new variable that will get us all the information contained in the "tr" brackets of the HTML from the website. 

```python
#find_content = soup.find_all('tr')
#print(find_content)
column_data = soup.find_all('tr')
#column_data
```

### For this part we create a nested FOR loop to get the information from the website's data table.

```python
for row in column_data[1:]: #For Loop that skips the first row (headers)
    row_data = row.find_all('td') #We get the information contained inside the td brackets
    individual_row_data = [data.text.strip() for data in row_data] # Creating a list using a nested for loop to get each element in the table
    #print(individual_row_data)
    length = len(df) #We get the length of the dataframe to add the information obtained to the end of the dataframe 
    df.loc[length] = individual_row_data # in order to not replace the existing one
```
