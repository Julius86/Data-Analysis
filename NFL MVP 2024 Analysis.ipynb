# Reading Files in Pandas

## Importing Pandas Library.

```python
import pandas as pd
```

## We will create a DataFrame by reading the first CSV file which has the QB Data. 

```python
df1 = pd.read_csv(r"C:\Users\julio.lopez1\Documents\Python Training\Jupyter Projects\CSV\NFLMVP2024-QB.csv")
df2 = pd.read_csv(r"C:\Users\julio.lopez1\Documents\Python Training\Jupyter Projects\CSV\NFLMVP2024-RB.csv")
df3 = pd.read_csv(r"C:\Users\julio.lopez1\Documents\Python Training\Jupyter Projects\CSV\NFLMVP2024-WR.csv")
```

```python
df1.info() #Show the information of the DataFrame
```

```python
df2.info()
```

```python
df3.info()
```

## We have to merge them all into a single DataFrame in order to be able to analyze it very easily.

```python
merge_df = df1.merge(df2, how='outer', on='Player') # We will merge all three Data Frames into one using an outer join.
full_df = merge_df.merge(df3, how='outer', on='Player')
full_df.fillna(0, inplace=True) # Replacing NaN values for zeroes (0)
```

```python
full_df.info() #Show the information of the full DataFrame.
```

```python
pd.set_option('display.max.rows',150) #As a way to check it is all complete, we show max rows and columns here.
pd.set_option('display.max.columns', 60)
```
